```python
from keras.models import Model
from keras.layers import *
```

```python
inputA = Input(shape=(128,))
inputB = Input(shape=(32,))
```

```python
x = Dense(64, activation='relu')(inputA)
hidden1 = Dense(32, activation='relu')(x)
hidden2 = Dense(4, activation='relu')(hidden1)
```

```python
y = Dense(8, activation='relu')(inputB)
hidden3 = Dense(4, activation='relu')(y)
```

```python
combined = concatenate([hidden2, hidden3])
```

```python
z = Dense(2, activation='relu')(combined)
z1 = Dense(1, activation='linear')(z)
```

```python
model = Model(inputs=[inputA, inputB], outputs=z1)
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model, show_shapes=True, to_file='model_multiple_inputs.png')
```
