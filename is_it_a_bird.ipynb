```python
urls = search_images_ddg('bird', max_images=1)
print(urls)
```

```python
dest = Path('bird.jpg')
if not dest.exists():
    download_url(urls[0], dest)
```

```python
img = Image.open(dest)
img.to_thumb(256,256)
```
