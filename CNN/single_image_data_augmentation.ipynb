```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img_1 = image.load_img('cat_img.jpg', target_size=(224, 224))
img_2 = image.load_img('dog_img.jpg', target_size=(224, 224))
type(img_1)
```

```python
import numpy as np

arr1 = image.img_to_array(img_1)
arr2 = image.img_to_array(img_2)
imgs_array = np.array([arr1, arr2])
type(imgs_array)
```

```python
imgs_array.shape
```

```python
datagen = ImageDataGenerator(
    rotation_range=40,      # Randomly rotate images up to 40 degrees
    width_shift_range=0.2,  # Fraction of total width to shift horizontally
    height_shift_range=0.2, # Fraction of total height to shift vertically
    shear_range=0.2,        # Shear intensity (tilt the image)
    zoom_range=0.2,         # Randomly zoom (in/out)inside pictures
    horizontal_flip=True,   # Randomly flip images horizontally
    vertical_flip=True,     # Randomly flip images vertically
    fill_mode='nearest'     # Strategy to fill newly created pixels after a shift or rotation
)

# Reshape the image to (number of batches, height, width, channels)
# img_array = img_array.reshape((1,) + img_array.shape)
# img_array.shape
```

```python
import os
if not os.path.exists('augmented_data'):
    os.makedirs('augmented_data')

i = 0
for batch in datagen.flow(
    imgs_array, 
    batch_size=2, 
    save_to_dir='augmented_data', 
    save_prefix='cat_dog', 
    save_format='jpg'
):
    i += 1
    if i > 20:
        break
```
