```python
from keras.applications.vgg16 import VGG16

model = VGG16()
model.summary()
```

```python
for i, layer in enumerate(model.layers):
    print(i, layer.name)
```

```python
filters, biases = model.layers[1].get_weights()
print(filters.shape)
print(biases.shape)
```

```python
f_min, f_max = filters.min(), filters.max()
filters = (filters - f_min) / (f_max - f_min)
```

```python
import matplotlib.pyplot as plt
n_filters, ix = 6, 1
for i in range(n_filters):
    f = filters[:, :, :, i]
    for j in range(3):
        ax = plt.subplot(n_filters, 3, ix)
        ax.set_xticks([])
        ax.set_yticks([])
        plt.imshow(f[:, :, j], cmap='gray')
        ix += 1
plt.show()
```

```python
from keras.models import Model

model = Model(inputs=model.inputs, outputs=model.layers[1].output)
```

```python
from keras.preprocessing.image import load_img, img_to_array
from keras.applications.vgg16 import preprocess_input
import numpy as np

img = load_img('othmane_img.jpeg', target_size=(224, 224))
img_array = img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = preprocess_input(img_array)
```

```python
features = model.predict(img_array)
print(features.shape)   
```

```python
fig = plt.figure(figsize=(15, 15))
for i in range(64):
    ax = fig.add_subplot(8, 8, i+1)
    ax.set_xticks([])
    ax.set_yticks([])
    plt.imshow(features[0, :, :, i], cmap='gray')
plt.show()
```

```python
model_2 = VGG16()
```

```python
layer_index = [ 2, 5 , 9 , 13 , 17]
outputs = [model_2.layers[i].output for i in layer_index]

model_3 = Model(inputs= model_2.inputs, outputs = outputs)
```

```python
feature_map = model_3.predict(img_array)

for i,fmap in zip(layer_index,feature_map):
    fig = plt.figure(figsize=(20,15))
    fig.suptitle("Layer_{}".format(i) , fontsize=20)
    for i in range(1,features.shape[3]+1):
        plt.subplot(8,8,i)
        plt.imshow(fmap[0,:,:,i-1] , cmap='gray')
    
plt.show()
```
