```python
import tensorflow
from tensorflow import keras
from keras.layers import Dense, Flatten, Conv2D, Input
from keras import Sequential
from keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
print(f"Training data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")
```

```python
model = Sequential(
    [
        Input(shape=(28, 28, 1)),
        Conv2D(32, (3, 3), activation='relu', padding='valid', strides=(2, 2)),
        Conv2D(32, (3, 3), activation='relu', padding='valid', strides=(1, 1)),
        Conv2D(32, (3, 3), activation='relu', padding='valid', strides=(1, 1)),
        Flatten(),
        Dense(128, activation='relu'),
        Dense(10, activation='softmax')
    ]
)
model.summary()
'''
Padding : valid means no padding, the convolution will only be applied 
to valid positions where the kernel fits entirely within the input image. 
This results in a smaller output feature map compared to the input.
same means that the convolution will be applied to all positions in the input image,
including the borders.

Stride : The stride parameter determines how much the convolutional kernel moves across the input image.
A stride of (1, 1) means that the kernel moves one pixel at a time both horizontally and vertically.
This results in a more detailed feature map but also increases the computational cost.
A larger stride (e.g., (2, 2)) means that the kernel moves two pixels at a time, which reduces
the spatial dimensions of the output feature map.
'''
```
